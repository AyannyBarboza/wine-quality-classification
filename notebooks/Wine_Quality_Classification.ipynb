Analisando a coorelação, identifcamos que alcohol apresenta a maior correlação positiva com a quality. Isso confirma a hipótese de que vinhos com maior graduação alcoólica tendem a receber notas mais altas dos avaliadores.

Já o volatile acidity apresenta uma forte correlação negativa com a quality, isso faz total sentido químico, pois a acidez volátil em excesso dá ao vinho um gosto de vinagre.


```python
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
variaveis_boxplot = ['alcohol', 'residual sugar', 'sulphates']


for i, coluna in enumerate(variaveis_boxplot):

    nome_real = coluna if coluna in df.columns else coluna.replace(' ', '_')

    if nome_real in df.columns:

        sns.boxplot(y=df[nome_real], ax=axes[i], color='#5F9EA0')
        axes[i].set_title(f'Boxplot de {coluna.title()}', fontsize=12, fontweight='bold')
        axes[i].set_ylabel('Valores')
    else:
        axes[i].set_title(f'Coluna "{coluna}" não encontrada')

plt.tight_layout()
plt.show()
```

Residual sugar apresenta uma quantidadealta de outliers na parte superior do gráfico.

Sulphates também apresenta um número considerável de outliers superiores.

Alcohol quase não apresenta outliers. Isso indica que o teor alcoólico dos vinhos segue uma faixa bem delimitada e uniforme.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(['quality', 'target'], axis=1, errors='ignore')
y = df['target']


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)


scaler = StandardScaler()


X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)


print(f"Dados de Treino (X_train): {X_train_scaled.shape} amostras e {X_train_scaled.shape[1]} colunas")
print(f"Dados de Teste (X_test): {X_test_scaled.shape} amostras e {X_test_scaled.shape[1]} colunas")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve

model_lr = LogisticRegression(random_state=42)
model_lr.fit(X_train_scaled, y_train)


model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train_scaled, y_train)


resultados = {}

for nome, modelo in [('Logistic Regression', model_lr), ('Random Forest', model_rf)]:

    y_pred = modelo.predict(X_test_scaled)

    y_prob = modelo.predict_proba(X_test_scaled)[:, 1]


    resultados[nome] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_prob)
    }


df_resultados = pd.DataFrame(resultados).T
print("=== TABELA DE AVALIAÇÃO DOS MODELOS ===")
display(df_resultados)


fig, axes = plt.subplots(1, 2, figsize=(14, 5))


for nome, modelo in [('Logistic Regression', model_lr), ('Random Forest', model_rf)]:
    y_prob = modelo.predict_proba(X_test_scaled)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    axes[0].plot(fpr, tpr, label=f'{nome} (AUC = {resultados[nome]["ROC-AUC"]:.2f})')

axes[0].plot([0, 1], [0, 1], 'k--', label='Chute Aleatório')
axes[0].set_title('Curva ROC - Comparação dos Modelos', fontweight='bold')
axes[0].set_xlabel('Taxa de Falso Positivo (1 - Especificidade)')
axes[0].set_ylabel('Taxa de Verdadeiro Positivo (Sensibilidade / Recall)')
axes[0].legend()


y_pred_rf = model_rf.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[1], cbar=False)
axes[1].set_title('Matriz de Confusão (Random Forest)', fontweight='bold')
axes[1].set_xlabel('Predição do Modelo')
axes[1].set_ylabel('Valor Real (Gabarito)')
axes[1].set_xticklabels(['Baixa/Média (0)', 'Alta (1)'])
axes[1].set_yticklabels(['Baixa/Média (0)', 'Alta (1)'])

plt.tight_layout()
plt.show()
```

O modelo Random Forest superou a Regressão Logística, com ROC-AUC de 0,91 e F1-Score de 0,69, demonstrando maior capacidade de discriminação e melhor desempenho em dados desbalanceados. A Random Forest é recomendada por capturar relações não lineares complexas entre as propriedades químicas do vinho, superando as limitações lineares da Regressão Logística.

```python
importancias = model_rf.feature_importances_
nomes_colunas = X.columns


df_importancia = pd.DataFrame({
    'Característica': nomes_colunas,
    'Importância': importancias
}).sort_values(by='Importância', ascending=False)


plt.figure(figsize=(10, 6))
sns.barplot(x='Importância', y='Característica', data=df_importancia, palette='viridis')


plt.title('Variáveis que Mais Influenciam a Qualidade do Vinho (Random Forest)', fontsize=14, fontweight='bold')
plt.xlabel('Grau de Importância', fontsize=12)
plt.ylabel('Propriedade Química', fontsize=12)
plt.tight_layout()
plt.show()
```

O gráfico de Feature Importance confirma as hipóteses levantadas na análise exploratória. O alcohol disparou como a variável de maior peso para a tomada de decisão da Random Forest.

Em seguida, propriedades como os sulphates e a volatile acidity também mostraram forte influência. Isso prova que o modelo foca em fatores químicos críticos relacionados ao paladar e à conservação para classificar a qualidade do vinho.
